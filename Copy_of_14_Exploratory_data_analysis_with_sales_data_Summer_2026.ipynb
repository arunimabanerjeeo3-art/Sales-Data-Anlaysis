### Question 2: Department Performance Consistency (3 Marks)

Which departments have the most stable weekly sales and which are the most volatile?

**Tasks:**
- Use standard deviation and coefficient of variation
- Identify highly stable and highly volatile departments

```python
# Calculating mean and standard deviation and coefficient of variation
#lowest cofficient of variation = Most stable
#Highest cofficient of variation = Most Volatile
department_cons = raw_df.groupby("department")["weekly_sales"].agg(['mean','std']).reset_index()
department_cons['CV'] = department_cons['std']/department_cons['mean']
department_cons_sorted = department_cons.sort_values(by='CV')
print(department_cons_sorted[["department","CV"]])



```

### Question 3: Holiday vs Non-Holiday Sales (3 Marks)

Analyze whether holidays significantly affect weekly sales.

**Tasks:**
- Compare average sales during holidays vs non-holidays
- Calculate percentage increase/decrease
- Identify departments benefiting most from holidays

```python
holiday_sales=(raw_df.groupby('is_holiday')['weekly_sales'].mean())
print(holiday_sales)
```

```python
non_holiday = holiday_sales[0]
holiday =holiday_sales[1]

percentage_increase = ((holiday - non_holiday)/non_holiday)*100

print("Percentage Increase:",percentage_increase)
```

```python
department_holiday = raw_df.groupby(['department','is_holiday'])['weekly_sales'].mean().unstack()
department_holiday['increase_%']=((department_holiday[1] - department_holiday[0])/department_holiday[0])*100

print(department_holiday.sort_values('increase_%',ascending=False))
```

### Question 4: Seasonal Trend Detection (3 Marks)

Identify whether sales increase or decrease during specific months.

**Tasks:**
- Extract month from the date column
- Compute average monthly sales
- Determine highest and lowest sales months

```python
raw_df['date'] = pd.to_datetime(raw_df['date'])
raw_df['month'] = raw_df['date'].dt.month_name()
monthly_sales = raw_df.groupby('month')['weekly_sales'].mean().reset_index()
highest_month = monthly_sales.loc[monthly_sales['weekly_sales'].idxmax()]
lowest_month = monthly_sales.loc[monthly_sales['weekly_sales'].idxmin()]

print(monthly_sales)

print("\nHighest Sales Month:")
print(highest_month)

print("\nLowest Sales Month:")
print(lowest_month)
```

### Question 5: Store and Department Contribution (4 Marks)

Which combinations of store and department contribute the most to overall revenue?

**Tasks:**
- Group using store_id and department
- Find top and worst-performing combinations
- Compute percentage contribution

```python
store_department = raw_df.groupby(['store_id','department'])['weekly_sales'].sum().reset_index()
top_combo = store_department.loc[store_department['weekly_sales'].idxmax()]
worst_combo = store_department.loc[store_department['weekly_sales'].idxmin()]
total_revenue = store_department['weekly_sales'].sum()
store_department['percentage_contribution']=(store_department['weekly_sales']/total_revenue*100)

print(store_department)
print("Top and worst performing combination")
print(top_combo)
print(worst_combo)
print("Percentage Contribution:")
print(store_department.head())
```

### Question 6: Load MNIST Data (3 Marks)

Load the MNIST dataset using `sklearn.datasets.load_digits`. Separate the dataset into features (`X`) and target labels (`y`).
Print the shape of the features and the target arrays.

**Expected Output:** The shape of `X` and `y`.

```python
from sklearn.datasets import load_digits
digits=load_digits()
x=digits.data
y=digits.target

print("Shape of x:",x.shape)
print("Shape of y:",y.shape)

```

### Question 7: K-Means Clustering (5 Marks)

Import `KMeans` from `sklearn.cluster`. Initialize and fit the K-Means clustering model on the MNIST features (`X`).
Since we know there are 10 digits (0-9), set the number of clusters to 10. Assign the cluster labels to a variable `kmeans_labels`.

**Expected Output:** A successfully fitted K-Means model and the cluster labels array.

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=10,random_state=42)
kmeans_labels = kmeans.fit_predict(x)

print("Cluster Labels:")
print(kmeans_labels)
```

### Question 8: F1 Score Evaluation for Clustering (5 Marks)

Evaluate the clustering performance using the F1 score. Since K-Means assigns arbitrary cluster labels, you will first need to map each cluster label to the most frequent true label in that cluster.
After mapping the labels, calculate and print the macro-averaged F1 score using `sklearn.metrics.f1_score`.

**Expected Output:** The calculated F1 score of the clustering.

```python
from sklearn.metrics import f1_score
from scipy.stats import mode
import numpy as np

label_mapping = {}
for cluster in np.unique(kmeans_labels):
  mask = (kmeans_labels == cluster)
  label_mapping[cluster] = mode(y[mask],keepdims=True).mode[0]

y_pred = np.array([label_mapping[label]for label in kmeans_labels])

f1=f1_score(y,y_pred,average='macro')

print("Macro F1 Score:",f1)
```
